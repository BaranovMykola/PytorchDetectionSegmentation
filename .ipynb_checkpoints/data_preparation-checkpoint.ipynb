```python
import pandas as pd
import os
import xml.etree.ElementTree as ET
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
trainval = './Data/CatVsDog/annotations/trainval.txt'
testval = './Data/CatVsDog/annotations/test.txt'
```

```python
def read_filenames(path):
    with open(path) as f:
        lines = f.readlines()

    names = [x.split(' ')[0] for x in lines]
    full_names = [os.path.join('./Data/CatVsDog/images/', x)+'.jpg' for x in names]
    return full_names, names
```

```python
train_files, train_names = read_filenames(trainval)
test_files, test_names = read_filenames(testval)
dataset = pd.DataFrame(data={'file': train_files+test_files,
                             'name': train_names+test_names,
                             'split': ['train']*len(train_files)+['test']*len(test_files)})
dataset.sample(4)
```

```python
dataset['mask_file'] = dataset.apply(lambda r: os.path.join('./Data/CatVsDog/annotations/trimaps/', r['name'])+'.png', axis=1)
dataset['annotation_file'] = dataset.apply(lambda r: os.path.join('./Data/CatVsDog/annotations/xmls/', r['name'])+'.xml', axis=1)
dataset.sample(5)
```

```python
files = dataset.apply(lambda r: os.path.isfile(r['file']), axis=1)
masks = dataset.apply(lambda r: os.path.isfile(r['mask_file']), axis=1)
labels = dataset.apply(lambda r: not os.path.isfile(r['annotation_file']), axis=1)
dataset.sample(5)
```

```python
root = None
def read_bb(row):
    global root
    if not os.path.isfile(row['annotation_file']):
        return None, None, None, None
    root = ET.parse(row['annotation_file']).getroot()
    xmin = int(next(root.iter('xmin')).text)
    ymin = int(next(root.iter('ymin')).text)
    xmax = int(next(root.iter('xmax')).text)
    ymax = int(next(root.iter('ymax')).text)

    return xmin, ymin, xmax, ymax

bbs = dataset.apply(read_bb, axis=1)
```

```python
dataset['xmin'] = [x[0] for x in bbs]
dataset['ymin'] = [x[1] for x in bbs]
dataset['xmax'] = [x[2] for x in bbs]
dataset['ymax'] = [x[3] for x in bbs]
dataset.sample(6)
```

```python
def visualize_sample(s):
    img = cv2.imread(s['file'])
    try:
        img = cv2.rectangle(img, (int(s['xmin']), int(s['ymin'])), (int(s['xmax']), int(s['ymax'])), (0,255,0), 3)
    except ValueError:
        print('No bounding box provided')
    finally:
        return img
    
img = visualize_sample(dataset.sample(1).iloc[0])
plt.imshow(img[:, :, ::-1])
```

```python
def get_img_shape(s):
    img = cv2.imread(s['file'])
    if img is None:
        return None, None
    
    return img.shape

shapes = dataset.apply(get_img_shape, axis=1)
```

```python
dataset['h'] = [x[0] for x in shapes]
dataset['w'] = [x[1] for x in shapes]
dataset['d'] = [x[2] for x in shapes]
dataset.sample(6)
```

```python
dataset['xmin_rel'] = dataset.apply(lambda r: r['xmin']/r['w'], axis=1)
dataset['xmax_rel'] = dataset.apply(lambda r: r['xmax']/r['w'], axis=1)
dataset['ymax_rel'] = dataset.apply(lambda r: r['ymax']/r['h'], axis=1)
dataset['ymin_rel'] = dataset.apply(lambda r: r['ymin']/r['h'], axis=1)
dataset['center_x'] = dataset.apply(lambda r: np.mean([r['xmin_rel'], r['xmax_rel']]), axis=1)
dataset['center_y'] = dataset.apply(lambda r: np.mean([r['ymin_rel'], r['ymax_rel']]), axis=1)
dataset['w_rel'] = dataset.apply(lambda r: r['xmax_rel']-r['xmin_rel'], axis=1)
dataset['h_rel'] = dataset.apply(lambda r: r['ymax_rel']-r['ymin_rel'], axis=1)
dataset.sample(6)
```

```python
def visualize_sample_rel(s):
    img = cv2.imread(s['file'])
    try:
        img = cv2.resize(img, (224, 224))
        pt1 = (int(s['xmin_rel']*img.shape[1]), int(s['ymin_rel']*img.shape[0]))
        pt2 = (int(s['xmax_rel']*img.shape[1]), int(s['ymax_rel']*img.shape[0])) 
        print(pt1, pt2)
        center = (int(s['center_x']*img.shape[1]), int(s['center_y']*img.shape[0])) 
        img = cv2.rectangle(img, pt1, pt2, (0,255,0), 3)
        img = cv2.circle(img, center, 6, (0,0,255), -1)
    except ValueError:
        print('No bounding box provided')
    finally:
        return img
    
img = visualize_sample_rel(dataset.iloc[34])
plt.imshow(img[:, :, ::-1])
```

```python
def visualize_sample_rel_center(s):
    img = cv2.imread(s['file'])
    try:
        img = cv2.resize(img, (224, 224))
        center = (int(s['center_x']*img.shape[1]), int(s['center_y']*img.shape[0])) 
        pt1 = (int((s['center_x']-s['w_rel']/2)*img.shape[1]), int((s['center_y']-s['h_rel']/2)*img.shape[0]))
        pt2 = (int((s['center_x']+s['w_rel']/2)*img.shape[1]), int((s['center_y']+s['h_rel']/2)*img.shape[0]))
        img = cv2.rectangle(img, pt1, pt2, (0,255,0), 3)
        img = cv2.circle(img, center, 6, (0,0,255), -1)
    except ValueError:
        print('No bounding box provided')
    finally:
        return img
    
img = visualize_sample_rel_center(dataset.iloc[34])
plt.imshow(img[:, :, ::-1])
```
